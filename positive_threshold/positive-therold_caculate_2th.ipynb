```python
import pandas as pd
import os
```

```python
reads_stats=pd.read_table('./Reads_stats.txt')
```

```python
dict_reads=reads_stats.set_index('Name').T.to_dict('list')
```

```python
dict_reads
```

```python
filenames=os.listdir('./positive+H2O/')
for filename in filenames:
    file=pd.read_table('./positive+H2O/'+filename)
    file_control=filename[0]+str(15)+'.txt'
    file1=pd.read_table('./control_CSF/'+file_control)
    print(filename)
    print(file_control)
    key=filename
    key=key[:-4]
    key1=key[0]
    new=key1+str(15)
    file_merge=file.merge(file1,left_on='NAME',right_on='NNAME',how='left')
    file_merge['RR/Reads']=file_merge['RR']/dict_reads[key]
    file_merge['NRR/Reads']=file_merge['NRR']/dict_reads[new]
    file_NAN=file_merge[file_merge['NRR'].isnull()]
    file_T=file_merge[file_merge['NRR'].notnull()]
    file_NAN['Real_ratio']=(file_NAN['RR']*dict_reads[new])/dict_reads[key]
    file_T['Real_ratio']=(file_T['RR']/dict_reads[key])/(file_T['NRR']/dict_reads[new])
    file_last=file_T.append(file_NAN,ignore_index=True)
    file_last=file_last.sort_values(by=['Real_ratio'],ascending=False)
    file_10_therold=file_last[file_last['Real_ratio']>=10]
    file_10_therold=file_10_therold.sort_values(by=['Real_ratio'],ascending=False)
    file_last.to_csv('./positive+negative/all/'+filename,sep='\t',index=False)
    file_10_therold.to_csv('./positive+negative/10_therold/'+filename,sep='\t',index=False)
```
