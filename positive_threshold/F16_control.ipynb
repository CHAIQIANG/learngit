```python
import pandas as pd
import os
```

```python
reads_stat=pd.read_table('./Reads_stats.txt')
```

```python
reads_stat
```

```python
dict_reads=reads_stat.set_index('Name').T.to_dict('list')
```

```python
dict_reads
```

```python
filenames=os.listdir('./positive+H2O/')
file_control=pd.read_table('./control_shui/F16.txt')
for filename in filenames:
    file=pd.read_table('./positive+H2O/'+filename)
    key=filename[:-4]
    file_merge=file.merge(file_control,left_on='NAME',right_on='NNAME',how='left')
    file_merge['RR/Reads']=file_merge['RR']/dict_reads[key]
    file_merge['NRR/Reads']=file_merge['NRR']/dict_reads['F16']
    file_NAN=file_merge[file_merge['NRR'].isnull()]
    file_T=file_merge[file_merge['NRR'].notnull()]
    file_NAN['Real_ratio']=(file_NAN['RR']*dict_reads['F16'])/dict_reads[key]
    file_T['Real_ratio']=file_T['RR/Reads']/file_T['NRR/Reads']
    file_last=file_T.append(file_NAN,ignore_index=True)
    file_10_therold=file_last[file_last['Real_ratio']>=10]
    file_last.to_csv('./F16_control/positive+H2O_done/'+filename,sep='\t',index=False)
    file_10_therold.to_csv('./F16_control/positive+H2O_10therold/'+filename,sep='\t',index=False)
```
